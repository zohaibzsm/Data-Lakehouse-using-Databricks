```python
import pyspark.sql.functions as F
from pyspark.sql.functions import trim, col, regexp_replace
```

### Load Silver Table

```python
df = spark.table("salesdatalakehouse.silver.erp_cust_az12")
```

### Invalid char check

```python
%sql
SELECT customer_id
FROM salesdatalakehouse.silver.erp_cust_az12
WHERE customer_id NOT IN (
SELECT customer_key
FROM salesdatalakehouse.silver.crm_cust_info)
```

## Invalid birthdate

```python
df.filter((col("birth_date") < F.to_date(F.lit("1926-01-01"), 'yyyy-MM-dd')) | (col("birth_date") > F.current_date())).show()
```
