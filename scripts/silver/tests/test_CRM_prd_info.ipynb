```python
import pyspark.sql.functions as F
from pyspark.sql.functions import trim, col, regexp_replace, substring, length
from pyspark.sql.types import StringType
```

### Load Silver Table

```python
df = spark.table("salesdatalakehouse.silver.crm_prd_info")
```

### Invalid char check

```python
df.filter(col("product_key").contains("-")).show()
```

##Normalization

```python
df.select(col('product_line')).distinct().show()
```

## Product Cost

```python
df.select("product_cost").filter(col("product_cost").isNull()).show()
```

##Dates

```python
(df
.filter(
    col("start_date") > col("end_date")
    ).orderBy("start_date").show())
```
