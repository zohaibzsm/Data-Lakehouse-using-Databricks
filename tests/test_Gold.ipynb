##Dim Customers Check

```python
%sql
SELECT 
    customer_key,
    COUNT(*) AS duplicate_count
FROM salesdatalakehouse.gold.dim_customers
GROUP BY customer_key
HAVING COUNT(*) > 1;
```

```python
%sql
SELECT DISTINCT gender 
-- all other columns were prepared and tested in silver layer
FROM salesdatalakehouse.gold.dim_customers
```

##Dim Products Check

```python
%sql
SELECT 
    product_key,
    COUNT(*) AS duplicate_count
FROM salesdatalakehouse.gold.dim_products
GROUP BY product_key
HAVING COUNT(*) > 1;
```

##Fact Table Check

```python
%sql
SELECT *
FROM salesdatalakehouse.gold.fact_sales s
LEFT JOIN salesdatalakehouse.gold.dim_customers c
ON s.customer_key = c.customer_key
LEFT JOIN salesdatalakehouse.gold.dim_products p
ON s.product_key = p.product_key
WHERE c.customer_key IS NULL OR p.product_key IS NULL;
```
