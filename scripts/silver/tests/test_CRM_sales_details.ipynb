```python
import pyspark.sql.functions as F
from pyspark.sql.functions import trim, col, regexp_replace, substring, length, abs
from pyspark.sql.types import StringType
```

### Load Silver Table

```python
df = spark.table("salesdatalakehouse.silver.crm_sales_details")
```

### Invalid char check

```python
df.filter(col("product_key").contains("-")).show()
```

##Sales and Price check

```python

# when sales is negative, zero or null, derive using quantity and price

df.filter(
            (col("sales_amount") <= 0 ) | 
            (col("sales_amount").isNull()) | 
            (col("sales_amount") != (col("quantity") * abs(col("price"))))
        ).show()
```

```python
# when price is negative, zero or null, derive using quantity and sales

df.filter(
            (col("price") <= 0 ) | 
            (col("price").isNull()) | 
            (col("price") != (col("sales_amount") / abs(col("quantity"))))
            ).show()
```
